```python
conn = pymssql.connect(server="133.186.215.165", database="HDCMART", user="hdcmart_user", password=password, charset="EUC-KR")
query = """
	SELECT
			CONCAT(YR_PLAN, RIGHT(MONTH,2)) AS YEARMONTH
		,	CD_SITE
		,	DS_SITE
		,	DS_ITEM
		,	DS_BONBU
		,	DS_SPNC_DEPT
		,	AM_PLAN
	FROM
		dbo.DATAMART_DIFT_PL_PLAN
		UNPIVOT (
			AM_PLAN
			FOR MONTH
			IN (AM_001, AM_002, AM_003, AM_004, AM_005, AM_006, AM_007, AM_008, AM_009, AM_010, AM_011, AM_012))
		AS UNPVT
	WHERE
		1=1
		AND CD_ITEM = 'PA000000'
		AND TY_GUBUN = 'A'
		AND TY_PLAN_GB = '01'
"""
data = pd.read_sql(query, conn)
```
